1. Create the below pattern using nested loop in Pyhton.


```python
for i in range(5):
    for j in range(i+1):
        print("*", end = " ")
    print("\r")
for k in range(5):
    for l in range(4-k):
        print("*", end = " ")
    print("\r")
```


2. Write a Python program to reverse a word after accepting the input from the user.

Input word: ineuron
Output: norueni

```python
st = input("Input Word:")
print("Output:",st[::-1])
```
