1.1) Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```python
def reduce(func, seq, specific_value_satrting=None): 
    generate = iter(seq) 
    if specific_value_satrting is None:
        value = next(generate) # it will generate next value of sequence
    else:
        value = specific_value_satrting
    for i in generate:
        value = func(value, i)
    return value
    
```

1.2) Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()

```python
def myfilter(fun,seq):
    empty = [] # creating empty list for storieng
    for i in seq:
        if fun(i): #passing sequence in function
            empty.append(i) #appending the empty list with desired output
    return empty #returning empty list
```

```python
# makeing chech even function
def check_even(cases):
    if cases%2 == 0:
        return True
    else:
        return False
```

```python
data = [4,6,8,4,45,46,654,5,9,468,5,48,54,6,8,]
```

```python
print(myfilter(check_even,data)) # testing myfilter function
```
